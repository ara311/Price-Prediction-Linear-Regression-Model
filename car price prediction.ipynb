# Model

```python
X=car.drop(columns='Price')
Y=car['Price']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split 
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder 
from sklearn.compose import make_column_transformer 
from sklearn.pipeline import make_pipeline
```

```python
ohe=OneHotEncoder()
ohe.fit(X[['Name','Label','Location','Kms_driven','Fuel_type','Owner','Year','Company']])
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Name','Label','Location','Kms_driven','Fuel_type','Owner','Year','Company']),remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_Train,Y_Train)
```

```python
Y_Pred=pipe.predict(X_Test)
Y_Pred
```

```python
r2_score(Y_Test,Y_Pred)
```

```python
scores=[]
for i in range(1000): 
    X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=0.2,random_state=i)
    lr=LinearRegression() 
    pipe=make_pipeline(column_trans,lr) 
    pipe.fit(X_Train,Y_Train) 
    Y_Pred=pipe.predict(X_Test)
    scores.append(r2_score(Y_Test,Y_Pred)) 
    
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_Train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=0.2,random_state=np.argmax(scores))
lr=LinearRegression() 
pipe=make_pipeline(column_trans,lr) 
pipe.fit(X_Train,Y_Train) 
Y_Pred=pipe.predict(X_Test) 
r2_score(Y_Test,Y_Pred)
```
